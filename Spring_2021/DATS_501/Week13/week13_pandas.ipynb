### "Groupby" in Pandas

```python
# read a dataset of alcohol consumption into a DataFrame
df = pd.read_csv('http://bit.ly/drinksbycountry')
df.head()
```

```python
# calculate the mean beer servings across the entire dataset
df.beer_servings.mean()
```

```python
# calculate the mean beer servings just for countries in Africa
df[df.continent=='Africa'].beer_servings.mean()
```

```python
# calculate the mean beer servings for each continent
df.groupby('continent').beer_servings.mean()
```

```python
# multiple aggregation functions can be applied simultaneously
df.groupby('continent').beer_servings.agg(['count', 'mean', 'min', 'max'])
```

```python
# specifying a column to which the aggregation function should be applied is not required
df.groupby('continent').mean()
```

```python
# allow plots to appear in the notebook
%matplotlib inline
```

```python
# side-by-side bar plot of the DataFrame directly above
df.groupby('continent').mean().plot(kind='bar')
```

### Select multiple rows and columns from a pandas DataFrame

```python
# read a dataset of UFO reports into a DataFrame
df = pd.read_csv('http://bit.ly/uforeports')
df.head(3)
```

* The <b> loc </b>  method is used to select rows and columns by <b> label. </b>  You can pass it:

    * A single label
    * A list of labels
    * A slice of labels
    * A boolean Series
    * A colon (which indicates "all labels")

```python
# row 0, all columns
df.loc[0, :]
```

```python
# rows 0 and 3 and 13, all columns
df.loc[[0, 3, 13], :]
```

```python
# rows 0 through 2 (inclusive), all columns
df.loc[0:2, :]
```

```python
# rows 0 through 2 (inclusive), columns 'City' and 'State'
df.loc[0:2, ['City', 'State']]
```

```python
# rows 0 through 2 (inclusive), columns 'City' through 'State' (inclusive)
df.loc[0:2, 'City':'State']
```

```python
# rows in which the 'City' is 'Oakland', column 'State'
df.loc[df.City=='Oakland', 'State']
```

### BONUS:  "lambda" and "map" function

```python
df= pd.read_csv('http://bit.ly/kaggletrain')
df.head()
```

* Let's find the distinct values in "Sex" column

```python
df.Sex.value_counts()
```

* Now, apply <b> Lambda/Map </b> functions to create a <b> dummy variable </b> for <b> Sex </b> column

```python
df['sex_lambda']= df['Sex'].apply(lambda x: 0 if x == 'female' else 1)
df.loc[888:889, :]
```

```python
df['sex_map']= df.Sex.map({'male': 0, 'female':1})
df.loc[888:889, :]
```
